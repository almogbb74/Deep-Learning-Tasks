### 1. Sigmoid Function Gradient

The sigmoid function is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

#### Task
1. Differentiate $\sigma(x)$ with respect to $x$.
2. Show that the derivative of $\sigma(x)$ can be expressed in terms of $\sigma(x)$ itself:
   
   $$
   \frac{d\sigma(x)}{dx} = \sigma(x)(1 - \sigma(x))
   $$

**Hint:** To simplify your work, start by applying the chain rule to the expression $\sigma(x) = \left(1 + e^{-x}\right)^{-1}$.

### 2. Tanh Function Gradient

The hyperbolic tangent function is defined as:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

#### Task
1. Differentiate $\tanh(x)$ with respect to $x$.
2. Show that the derivative of $\tanh(x)$ can be expressed as:

   $$
   \frac{d\tanh(x)}{dx} = 1 - \tanh(x)^2
   $$

**Hint:** You may want to express $\tanh(x)$ in terms of $\sigma(x)$, noting that $\tanh(x) = 2\sigma(2x) - 1$, as an alternative approach.


# Solution


**Task 1.a: Differentiate $\sigma(x)$:**

We'll use the hint and start with the form $\sigma(x) = (1 + e^{-x})^{-1}$.
We will use the **Chain Rule**: $\frac{d}{dx}f(g(x)) = f'(g(x)) \cdot g'(x)$

1.  **Identify outer and inner functions:**
    * Outer function: $f(u) = u^{-1}$ (For clarifaction $u$ is a placeholder)
    * Inner function: $g(x) = 1 + e^{-x}$

2.  **Find the derivatives of each part:**
    * $f'(u) = -1 \cdot u^{-2} = -\frac{1}{u^2}$
    * $g'(x) = \frac{d}{dx}(1 + e^{-x}) = 0 + e^{-x} \cdot (-1) = -e^{-x}$

3.  **Apply the Chain Rule:**
    $$\frac{d\sigma}{dx} = f'(g(x)) \cdot g'(x)$$
    $$\frac{d\sigma}{dx} = \left( -\frac{1}{(1 + e^{-x})^2} \right) \cdot ( -e^{-x} )$$

4.  **Simplify the expression:**
    The two negatives cancel out, leaving us with the final derivative.
    $$\frac{d\sigma}{dx} = \frac{e^{-x}}{(1 + e^{-x})^2}$$

---

**Task 1.b: Show $\frac{d\sigma}{dx} = \sigma(x)(1 - \sigma(x))$:**

We will prove this by starting with the right-hand side, $\sigma(x)(1 - \sigma(x))$, and showing that it simplifies to our result from Task 1.

1.  **Start with the expression:**
    $$\sigma(x)(1 - \sigma(x))$$

2.  **Substitute the definition of $\sigma(x) = \frac{1}{1 + e^{-x}}$:**
    $$= \left( \frac{1}{1 + e^{-x}} \right) \cdot \left( 1 - \frac{1}{1 + e^{-x}} \right)$$

3.  **Simplify the second term** by finding a common denominator:
    $$1 - \frac{1}{1 + e^{-x}} = \frac{(1 + e^{-x})}{(1 + e^{-x})} - \frac{1}{1 + e^{-x}}$$
    $$= \frac{1 + e^{-x} - 1}{1 + e^{-x}}$$
    $$= \frac{e^{-x}}{1 + e^{-x}}$$

4.  **Substitute this simplified term back** into our main expression:
    $$= \left( \frac{1}{1 + e^{-x}} \right) \cdot \left( \frac{e^{-x}}{1 + e^{-x}} \right)$$

5.  **Multiply the two fractions:**
    $$= \frac{1 \cdot e^{-x}}{(1 + e^{-x})(1 + e^{-x})}$$
    $$= \frac{e^{-x}}{(1 + e^{-x})^2}$$

This final expression is identical to the derivative we calculated in Task 1. Thus, we have shown that:
$$\frac{d\sigma(x)}{dx} = \sigma(x)(1 - \sigma(x))$$

**Task 2.a: Differentiate $\tanh(x):$**


We start with the definition:
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

We use the **Quotient Rule**: $\left(\frac{g}{h}\right)' = \frac{g'h - gh'}{h^2}$

Let:
* $g(x) = e^x - e^{-x} \implies g'(x) = e^x + e^{-x}$
* $h(x) = e^x + e^{-x} \implies h'(x) = e^x - e^{-x}$

Plug into the rule:
$$\frac{d}{dx}\tanh(x) = \frac{(e^x + e^{-x})(e^x + e^{-x}) - (e^x - e^{-x})(e^x - e^{-x})}{(e^x + e^{-x})^2}$$
Simplify:
$$\frac{d}{dx}\tanh(x) = \frac{(e^x + e^{-x})^2 - (e^x - e^{-x})^2}{(e^x + e^{-x})^2}$$

Simplify the numerator:
* $(e^{2x} + 2 + e^{-2x}) - (e^{2x} - 2 + e^{-2x}) = 4$

So, the derivative is:
$$\frac{d}{dx}\tanh(x) = \frac{4}{(e^x + e^{-x})^2}$$

---

**Task 2.b: Show $\frac{d}{dx}\tanh(x) = 1 - \tanh(x)^2$:**


We will work with the right-hand side:
$$1 - \tanh(x)^2 = 1 - \left(\frac{e^x - e^{-x}}{e^x + e^{-x}}\right)^2$$

Find a common denominator:
$$= \frac{(e^x + e^{-x})^2}{(e^x + e^{-x})^2} - \frac{(e^x - e^{-x})^2}{(e^x + e^{-x})^2}$$

Combine the fractions:
$$= \frac{(e^x + e^{-x})^2 - (e^x - e^{-x})^2}{(e^x + e^{-x})^2}$$

As we found in Task 1, the numerator simplifies to 4:
$$= \frac{4}{(e^x + e^{-x})^2}$$

This result is identical to the derivative we calculated in Task 1.
Therefore, $\frac{d}{dx}\tanh(x) = 1 - \tanh(x)^2$.
